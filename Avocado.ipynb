```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv('avocado.csv',parse_dates=['Date','year'],index_col=['Date'])
data=data.drop(['Unnamed: 0','year'],axis=1)
data.sort_index()
```

```python
dataDum=pd.get_dummies(data['type'])
dataDum
```

```python
dataDum1=pd.get_dummies(data['region'])
dataDum1
```

```python
dataFinal=pd.concat([data,dataDum,dataDum1],axis=1)
dataFinal
```

```python
y=dataFinal['AveragePrice']
x=dataFinal.drop(['AveragePrice','type','region'],axis='columns')
x
```

```python
corr=dataFinal.corr()
plt.figure(figsize=(50,20))
plt.imshow(corr,cmap='gray')
plt.yticks(np.arange(len(dataFinal.columns)),dataFinal.columns)
plt.xticks(np.arange(len(dataFinal.columns)),dataFinal.columns,rotation='vertical')
plt.colorbar()
# sb.heatmap(corr)
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.score(x,y)*100
```

```python

k=([[64236.62,1036.74,54454.85,48.16,8696.87,8603.62,93.25,0.00,1,0]])
x='Atlanta'
for item in range(len(dataDum1.columns)):
    if x==dataDum1.columns[item]:
        k[0].append(1)
    else:
        k[0].append(0)
model.predict(k)
```
