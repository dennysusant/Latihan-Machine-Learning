```python
import numpy as np 
import pandas as pd 
```

```python
datarating=pd.read_csv('rating.csv')
datarating
```

```python
dfRating=pd.DataFrame(
    datarating[:100000].drop('timestamp',axis=1)
)
dfRating
```

```python
dfPivot=dfRating.pivot_table(
    index=['userId'],columns=['movieId'],values='rating'
)
dfPivot=dfPivot.fillna(0)
dfPivot.head()
```

```python
corrMatrix=dfPivot.corr()
corrMatrix
```

```python
userX=[(1,5),(2,5),(13,5),(48,4)]
finalSkor=pd.DataFrame()
for movieId,rating in userX:
    skor=corrMatrix[movieId]*rating
    skor=skor.sort_values(ascending=False)
    finalSkor=finalSkor.append(skor)
finalSkor=finalSkor.sum().sort_values(ascending=False)
finalSkor[:15].index.values
```

```python
dfMovie=pd.read_csv('movie.csv')
listMovid=finalSkor[:15].index.values
for i in listMovid:
    print(dfMovie[dfMovie['movieId']==i]['title'])
```
